```python
import sqlite3
import pandas as pd
```

```python
con = sqlite3.connect('/content/book_simple.db')
cur = con.cursor() 
```

```python
# Выведите авторов, имеющихся в базе
name = """
SELECT name
FROM author;
"""
```

```python
pd.read_sql(name, con)
```

```python
# Найдите всех авторов с инициалами "С. А.". Автор в базе определяется так: И. О. Фамилия
```

```python
c_a = """
SELECT name
FROM author
Where name like "С. А.%";
"""
```

```python
pd.read_sql(c_a, con)
```

```python
#Выведите список книг, выпущенных в издательстве "Мир".
```

```python
mir = """
SELECT book.title, 
       publisher.name
FROM book
INNER JOIN publisher
ON book.publisher_id = publisher.id
WHERE publisher.name like 'Мир';
"""
```

```python
pd.read_sql(mir, con)
```

```python
# Выведите количество книг, принадлежащих каждой из категорий. Результат отсортировать по возрастанию.
```

```python
cat_counts = """
SELECT category.title,
       COUNT(book.title) AS book_counts
FROM book
INNER JOIN category
ON book.category_id = category.id
GROUP BY category.title
ORDER BY book_counts ASC 
"""
```

```python
pd.read_sql(cat_counts, con)
```

```python
# Выведите количество книг, выпущенных в каждом издательстве. В итоговом списке оставьте только те издательства, которые выпустили 3 и более книг. Результат отсортировать по убыванию.
```

```python
publish_name = """
SELECT publisher.name, COUNT(book.title) AS book_coutns
FROM book
INNER JOIN publisher
ON publisher.id = book.publisher_id
GROUP BY publisher.name
HAVING book_coutns > 3 
ORDER BY book_coutns DESC  
"""
```

```python
pd.read_sql(publish_name, con)
```

```python
# Выведите список книг, автором которых являлся <<С. А. Айвазян>>.
```

```python
ayvasyan = """
SELECT title, author.name as author
FROM book
INNER JOIN authorbook
ON book.id = authorbook.book_id
INNER JOIN author
ON author.id = authorbook.author_id
WHERE name like "С. А. Айвазян";
"""
```

```python
pd.read_sql(ayvasyan, con)
```

```python
# В каких издательствах и сколько выпускались книги за авторством <<С. А. Айвазян>>. Результат отсортировать по убыванию.
```

```python
publisher_ayvasyan = """
SELECT publisher.name, author.name, COUNT(book.title) AS book_count
FROM publisher
INNER JOIN book
ON publisher.id = book.publisher_id
INNER JOIN authorbook
ON book.id = authorbook.book_id
INNER JOIN author
ON author.id = authorbook.author_id
GROUP BY author.name
HAVING author.name like "С. А. Айвазян";
"""
```

```python
pd.read_sql(publisher_ayvasyan, con)
```
